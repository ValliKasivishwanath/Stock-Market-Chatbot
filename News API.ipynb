## Last 7 Days News 

```python
import requests
import datetime
import pandas as pd
import time

# Your Finnhub API Key
FINNHUB_API_KEY = 'd0p67q9r01qr8ds1p4tgd0p67q9r01qr8ds1p4u0'

# Set date range (past 3 days)
today = datetime.datetime.now()
seven_days_ago = today - datetime.timedelta(days=7)
from_date = seven_days_ago.strftime('%Y-%m-%d')
to_date = today.strftime('%Y-%m-%d')

# Collect news
all_news = []

for idx, ticker in enumerate(tickers):
    url = f'https://finnhub.io/api/v1/company-news?symbol={ticker}&from={from_date}&to={to_date}&token={FINNHUB_API_KEY}'
    response = requests.get(url)

    try:
        data = response.json()
    except Exception as e:
        print(f"[{idx+1}/{len(tickers)}] Failed to parse JSON for {ticker}: {e}")
        continue

    if isinstance(data, list):  # Normal case: API returned articles list
        for article in data:
            if isinstance(article, dict):
                news_item = {
                    'ticker': ticker,
                    'headline': article.get('headline', ''),
                    'summary': article.get('summary', ''),
                    'datetime': datetime.datetime.fromtimestamp(article.get('datetime', 0)),
                    'url': article.get('url', '')
                }
                all_news.append(news_item)
            else:
                print(f"[{idx+1}/{len(tickers)}] Skipped non-dict article for {ticker}")
    else:
        print(f"[{idx+1}/{len(tickers)}] Non-list response for {ticker}: {data}")

    print(f"[{idx+1}/{len(tickers)}] Completed {ticker}")
    time.sleep(1)  # Increase wait time to avoid hitting rate limit

# Save to CSV
df = pd.DataFrame(all_news)
df.to_csv('News_Data.csv', index=False)
print("âœ… News collection complete.")
```

## Transfer News Data from python to SQL

```python
import pandas as pd
import mysql.connector

# Load your CSV file
news_data = pd.read_csv('/Users/ashwin/Desktop/Stock Market Chatbot/News_Data.csv')

# Convert datetime column to string format suitable for MySQL
news_data['datetime'] = pd.to_datetime(news_data['datetime'])
news_data['datetime'] = news_data['datetime'].dt.strftime('%Y-%m-%d %H:%M:%S')

# Connect to MySQL
conn = mysql.connector.connect(
    host='localhost',
    user='root',
    password='Vallik271297*',
    database='stock_market'
)
cursor = conn.cursor()



# Insert data row by row
for _, row in news_data.iterrows():
    sql = """
    INSERT INTO news (ticker, headline, summary, datetime, url)
    VALUES (%s, %s, %s, %s, %s)
    """
    values = (
        row['ticker'],
        row['headline'],
        row['summary'],
        row['datetime'],
        row['url']
    )
    cursor.execute(sql, values)

conn.commit()
print("âœ… News data with sentiment inserted into MySQL.")
cursor.close()
conn.close()
```
