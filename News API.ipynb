## Pull the Last one Day News 

```python
import os
import datetime
import requests
import pandas as pd
import time

FINNHUB_API_KEY = 'd0p67q9r01qr8ds1p4tgd0p67q9r01qr8ds1p4u0'  
csv_path = 'News_Data.csv'

# Get tickers
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
tickers = pd.read_html(url)[0]['Symbol'].str.replace('.', '-', regex=False).tolist()

# 1ï¸âƒ£ Check existing data
if os.path.exists(csv_path):
    df_existing = pd.read_csv(csv_path)
    df_existing['datetime'] = pd.to_datetime(df_existing['datetime'])
    latest_datetime = df_existing['datetime'].max()
    print(f"âœ… Latest news in your file: {latest_datetime}")
else:
    latest_datetime = datetime.datetime.now() - datetime.timedelta(days=1)
    df_existing = pd.DataFrame()
    print(f"âš ï¸ No existing file found. Pulling last 1 day's news from {latest_datetime}")

# 2ï¸âƒ£ Set date range for pull
start_datetime = latest_datetime
end_datetime = datetime.datetime.now()
print(f"ðŸ“… Pulling news from {start_datetime} to {end_datetime}")

# 3ï¸âƒ£ Pull news
all_news = []

for idx, ticker in enumerate(tickers):
    url = f'https://finnhub.io/api/v1/company-news?symbol={ticker}&from={start_datetime.date()}&to={end_datetime.date()}&token={FINNHUB_API_KEY}'
    response = requests.get(url)

    if response.status_code != 200:
        print(f"[{idx+1}/{len(tickers)}] API error for {ticker}: Status {response.status_code}")
        continue

    try:
        data = response.json()
    except Exception as e:
        print(f"[{idx+1}/{len(tickers)}] JSON parse error for {ticker}: {e}")
        continue

    if isinstance(data, list):
        for article in data:
            article_time = datetime.datetime.fromtimestamp(article.get('datetime', 0))
            if article_time > latest_datetime:
                news_item = {
                    'ticker': ticker,
                    'headline': article.get('headline', ''),
                    'summary': article.get('summary', ''),
                    'datetime': article_time,
                    'url': article.get('url', '')
                }
                all_news.append(news_item)

    print(f"[{idx+1}/{len(tickers)}] Done {ticker}")
    time.sleep(1)  # To respect API limits

# 4ï¸âƒ£ Save combined results
if all_news:
    df_new = pd.DataFrame(all_news)
    df_combined = pd.concat([df_existing, df_new], ignore_index=True)
    df_combined.drop_duplicates(subset=['ticker', 'datetime', 'headline'], inplace=True)
    df_combined.to_csv(csv_path, index=False)
    print(f"âœ… Updated CSV with {len(df_new)} new articles (after dedup: {len(df_combined)} total).")
else:
    print("âš ï¸ No new articles found since last pull.")
```
