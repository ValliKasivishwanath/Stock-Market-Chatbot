```python
import pandas as pd

# Path to your uploaded file
file_path = '/Users/ashwin/Desktop/Stock Market Chatbot/News_Data.csv'

# Read the CSV into a pandas DataFrame
news_data = pd.read_csv(file_path)

news_data
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize VADER
analyzer = SentimentIntensityAnalyzer()

# Define sentiment function
def get_sentiment(text):
    score = analyzer.polarity_scores(text)['compound']
    if score > 0.1:
        return "Positive"
    elif score < -0.1:
        return "Negative"
    else:
        return "Neutral"

# Combine headline + summary for better context
news_data['text_for_sentiment'] = news_data['headline'].fillna('') + ' ' + news_data['summary'].fillna('')

# Apply sentiment
news_data['Sentiment'] = news_data['text_for_sentiment'].apply(get_sentiment)

news_data[['ticker','headline','Sentiment']]
```

```python
grouped = news_data.groupby('Sentiment')['ticker'].count()
grouped
```
